Q2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of

Q4. Write a python program to scrap book name, author name, genre and book review of any 5 books from 
‘www.bookpage.com’

```python
page4 = requests.get('https://bookpage.com/reviews/')
soup4 = BeautifulSoup(page4.content, 'html.parser')
```

```python
url_tags = soup4.find_all("h4",class_="italic")
```

```python
urls = []
for i in url_tags:
    for j in i.find_all("a", href=True):
        if j.text:
            urls.append(j['href'])
```

```python
book_name = []
```

```python
author_name = []
```

```python
book_genre = []
```

```python
books_review = []
```

```python
for url in urls:
        book = requests.get('https://www.bookpage.com'+url)
        soup = BeautifulSoup(book.content, 'html.parser')
        book_name.append(soup.find('h1').text.replace('\n','').replace('★',''))  
        author_name.append(soup.find('h4').text.replace('\n',''))  
        book_genre.append(soup.find('p', class_="genre-links").text.replace('\n',''))   
        books_review.append(soup.find('div', class_= "article-body").text.replace('\n',''))  
```

```python
books=pd.DataFrame({})
books["Book Name"]=book_name[:5]
books["Author"]=author_name[:5]
books["Genre"]=book_genre[:5]
books["Review"]=books_review[:5]
books
```

from the above we have taken 5 books with name, author name, genre and review.



Q5. Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:
    
    
    (i)  Top 10 ODI teams in men’s cricket along with the records for matches, points and rating. 

```python
url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
page5 = requests.get(url)
```

```python
soup5 = BeautifulSoup(page5.content)
```

```python
team = soup5.find_all("span",class_='u-hide-phablet')
team_name = []
for i in team:
    team_name.append(i.text)
```

```python
matches = [] 
points = [] 
ratings = [] 
new_list = [] 
```

```python
for i in soup5.find_all("td",class_='rankings-block__banner--matches'): 
    matches.append(i.text)
for i in soup5.find_all("td",class_='rankings-block__banner--points'):
    points.append(i.text)
for i in soup5.find_all("td",class_='rankings-block__banner--rating u-text-right'):
    ratings.append(i.text.replace("\n",""))
for i in soup5.find_all("td",class_='table-body__cell u-center-text'):
    new_list.append(i.text)
for i in range(0,len(new_list)-1,2):
    matches.append(new_list[i]) 
    points.append(new_list[i+1]) 
for i in soup5.find_all("td",class_='table-body__cell u-text-right rating'):
    ratings.append(i.text)
```

```python
icc=pd.DataFrame({})
icc['Team_name']=team_name[:10]
icc['Matches']=matches[:10]
icc['Points']=points[:10]
icc['Ratings']=ratings[:10]
icc
```

As we can see the top 10 men's cricket team with Team name, points , matches and ratings also.

Q7. Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The 
scraped data should include Product Name, Price, Image URL and Average Rating.

```python
#send get request to the webpage server to get the source code of the page
url = "https://www.amazon.in/s?k=Mobile+phones+under+20000&ref=nb_sb_noss_2"
page11 = requests.get(url)
#see content in page11
soup11 = BeautifulSoup(page11.content)
product_name = []  #empty list
price = []    #empty list
rating = []   #empty list
img_url = []   #empty list


# Scrape product name
for i in soup11.find_all("span",class_="a-size-medium a-color-base a-text-normal"):
    product_name.append(i.text)

# Scrape product price
for i in soup11.find_all("span",class_="a-price-whole"):
    price.append(i.text)
for i in soup11.find_all("span",class_="a-icon-alt"):
    rating.append(i.text)

# Scrape images url
for i in soup11.find_all("img",class_="s-image"):
    img_url.append(i.get("src"))
    
# Make data frame mobile phones under Rs. 20,000 listed on Amazon.in
mobile_phones=pd.DataFrame({})
mobile_phones['product_name']=product_name[:12]
mobile_phones['price']=price[:12]
mobile_phones['Rating']=rating[:12]
mobile_phones['img_url'] = img_url[:12]
mobile_phones
```

In this we can see datail  report of mobile phones with product name, price, rating and img_url.

Q8. Write a python program to extract information about the local weather from the National Weather Service 
website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day 
extended forecast display for the city. The data should include period, short description, temperature and 
description. 

```python

url = "https://forecast.weather.gov/MapClick.php?textField1=37.78&textField2=-122.42#.YJJVsLUzZPY"
page12 = requests.get(url)

soup12 = BeautifulSoup(page12.content)
period = []  
short_desc = []  
temperature = []   
description = []    


for i in soup12.find_all("div", class_="col-sm-2 forecast-label"):
    period.append(i.text)
    

for i in soup12.find_all('div',class_="col-sm-10 forecast-text"):        
        short_desc.append(i.text.split(',')[0])
for i in soup12.find_all('div',class_="col-sm-10 forecast-text"):        
        description.append(i.text)


for i in soup12.find_all('p',attrs={'short-desc'}):
        if i.next_sibling is not None:
            temperature.append(i.next_sibling.text)
        else:
            temperature.append(' ')


for i in soup12.find_all("div", class_="col-sm-10 forecast-text"):
        description.append(i.text)

san_francisco_weather=pd.DataFrame({})
san_francisco_weather['Period']=period[:9]
san_francisco_weather['Temperature']=temperature[:9]
san_francisco_weather['Short_description']=short_desc[:9]
san_francisco_weather['Description']=description[:9]
san_francisco_weather
```

in this we can see the weather report period wise,temp. wise, description wise.

Q9. Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, 
company name, CTC, and apply date.

```python
url = "https://internshala.com/fresher-jobs"
page13 = requests.get(url)
```

```python
soup13 = BeautifulSoup(page13.content)
job_titles = []  
company_names = []   
CTC = []   
apply_date = []   
```

```python
titles=soup13.find_all('div',class_="heading_4_5 profile") 
for i in titles:
    job_titles.append(i.text.replace('\n', ''))
    
    
companies=soup13.find_all('a',class_="link_display_like_text")
for i in companies:
    company_names.append(i.text.strip())



job_info = []
info = soup13.find_all("div",class_="item_body")
for i in info:
    job_info.append(i.text.replace("\n","").replace("\xa0",""))
    
    
for i in range(2,len(job_info),3):
    apply_date.append(job_info[i].strip())
    
for i in range(1,len(job_info),3):
    CTC.append(job_info[i].strip())
    

JOB=pd.DataFrame({})
JOB['Job_Titles']=job_titles
JOB['Company_Names']=company_names
JOB['CTC']=CTC
JOB['Apply_Date']=apply_date
JOB
```

in this we can see the job opening for freshers with company name,job titles, ctc and apply date.
